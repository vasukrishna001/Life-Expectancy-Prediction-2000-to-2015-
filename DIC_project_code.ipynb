```python
#to visualize the top 10 and least ten countries based on avg of life expentancy,adult mortality , infant deadths since 2000-2015
df_tencountries = df[(df['Year'] >= 2000) & (df['Year'] <= 2015)]

# Grouping by countries with their means of 3 columns as mentioned
mean_val = df_tencountries.groupby('Country').agg({
    'Life_Expectancy': 'mean',
    'Adult_Mortality': 'mean',
    'Infant_Deaths': 'mean'
    }).reset_index()

# Function to plot the first and last 10 rows of data based on averages
def tencountries(mean_val, col_name):
    # Sorting and selecting the top and least rows
    top_10 = mean_val.sort_values(by=col_name, ascending=False).head(10)
    bottom_10 = mean_val.sort_values(by=col_name, ascending=True).head(10)
    #plotting
    plt.figure(figsize=(20, 10))
    #first top 10
    plt.subplot(1, 2, 1)
    sns.barplot(x=top_10['Country'], y=top_10[col_name], color='Blue')
    plt.xticks(rotation=90,fontweight='bold')
    plt.title(f'Top 10 Countries by AVG of {col_name}(2000-2015)',fontsize=20, fontweight='bold')
    # least 10
    plt.subplot(1, 2, 2)
    sns.barplot(x=bottom_10['Country'], y=bottom_10[col_name], color='Blue')
    plt.xticks(rotation=90,fontweight='bold')
    plt.title(f'Least 10 Countries by AVG of {col_name}(2000-2015)',fontsize=20, fontweight='bold')
    plt.tight_layout()
    plt.show()

# Visualizing
tencountries(mean_val, 'Life_Expectancy')
tencountries(mean_val, 'Adult_Mortality')
tencountries(mean_val, 'Infant_Deaths')
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred_lr)
print(f"Mean Absolute Error (MAE): {mae}")
```

#### Plotting MSE for all the modles

```python
import matplotlib.pyplot as plt
# Extracting model_list and their corresponding MSE values
model_list= list(final_scores.keys())
MSE_list = []
for i in model_list:
    MSE_value = final_scores[i]['MSE']
    MSE_list.append(MSE_value)
# Generating labels with corresponding MSE values
lab_list = []
for i,j in zip(model_list, MSE_list):
    label = f'{i}\n{j:.2f}'
    lab_list.append(label)

# pie chart creation
plt.figure(figsize=(6, 4))
plt.pie(MSE_list, labels=lab_list, startangle=100, shadow=True, explode=(0, 0.2, 0, 0, 0, 0))
# Adding white circle at middle
figure = plt.gcf()
figure.gca().add_artist(plt.Circle((0, 0), 0.50, fc='white'))
# Plotting
plt.title('MSE of All 6 Different Models', fontsize=14, color='blue')
plt.show()
```

#### Plotting MAE for all the modles

```python
# Extracting models and their corresponding MAE values
MAE_list = []
for i in model_list:
    MAE_val = final_scores[i]['MAE']
    MAE_list.append(MAE_val)
# Generating labels with corresponding MAE values
# Generating labels with corresponding MAE values in a different way
lab_list = []
for i,j in zip(model_list, MAE_list):
    label = f'{i}\n{j:.2f}'
    lab_list.append(label)
# pie chart creation
plt.figure(figsize=(6, 4))
plt.pie(MAE_list, labels=lab_list, startangle=100, shadow=True, explode=(0, 0.2, 0, 0, 0, 0))
# Adding white circle at middle
figure = plt.gcf()
figure.gca().add_artist(plt.Circle((0, 0), 0.50, fc='white'))
# Plotting
plt.title('MAE of ALL 6 Different Models', fontsize=16, color='blue')
plt.show()
```

#### Plotting r^2 for all the modles

```python
# Extracting models and their corresponding  r^2 values
r_square_list = []
for i in model_list:
    r2_value = final_scores[i]['R2']
    r_square_list.append(r2_value)
# Generate labels with R² values
labs_list = []
for i, j in zip(model_list, r_square_list):
    label = f'{i}\n{j:.2f}'
    labs_list.append(label)
# Create the pie chart
plt.figure(figsize=(6, 4))
plt.pie(r_square_list, labels=labs_list, startangle=100, shadow=True, explode=(0, 0.2, 0, 0, 0, 0))
# Adding white circle at middle
figure = plt.gcf()
figure.gca().add_artist(plt.Circle((0, 0), 0.50, fc='white'))
# Plotting
plt.title('R² of all 6 Different Models', fontsize=14, color='blue')
plt.show()
```
